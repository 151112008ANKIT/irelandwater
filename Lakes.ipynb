```python
#Intersection between River and Lakes
Rivers_merged_lake=gpd.overlay(riv,lak,how='intersection')
Rivers_merged_lake.plot(figsize=(10,10),cmap='jet',column='PERIMETER')
```

```python
Rivers_merged_lake['Area(km2)']=Rivers_merged_lake.area/1000
```
